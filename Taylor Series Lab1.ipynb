# ⚓ First **Function:**: $f(x) = e^x$

Taylor Series Expansion Around $a = 0$:

$f(x) = \frac{f^{(0)}(0)}{0!}x^0 + \frac{f^{(1)}(0)}{1!}x^1 + \frac{f^{(2)}(0)}{2!}x^2 + \frac{f^{(3)}(0)}{3!}x^3 + ...$

Substituting the derivatives evaluated at $a = 0$:

$f(x) = \frac{e^0}{0!}x^0 + \frac{e^0}{1!}x^1 + \frac{e^0}{2!}x^2 + \frac{e^0}{3!}x^3 + ...$

Since $e^0 = 1$, we can simplify:

$f(x) = \frac{1}{0!}x^0 + \frac{1}{1!}x^1 + \frac{1}{2!}x^2 + \frac{1}{3!}x^3 + ...$

This can be written in summation notation as:

# $f(x) = \sum_{n=0}^{\infty}  \frac{x^n}{n!}$



# **Plotting the exponential nonlinear function and its approximations**



```python
import numpy as np
import matplotlib.pyplot as plt
import math

## to get an array with evenly spaced values. (size = 200)
x = np.linspace(-np.pi, np.pi, 200)  # هنا قيمة x مجموعة ارقام لاني ابغا ارسم رسمة متوسعة عكس تحت اخترت نقطة معينة
y = np.zeros(len(x)) ## initialize all y for every x above to 0 to prepare the sum (size= same as x)
plt.figure(figsize = (8,6)) ## setting the size of the plot

# loop to calculate the taylor series
for n in range(8):
    y = y + (x ** n) / math.factorial(n)
    plt.plot(x, y, label=f'Degree={n}')
# نرسم الصح
plt.plot(x, np.exp(x), 'k', label='Analytic value', linewidth=2)


plt.grid()
plt.title('Taylor Series Approximations of e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()  # زي مفتاح الخريطة
plt.show()
```



> We see that once we increase the degree we come closer to the true value::



# Comparing analytical and approximated value of $f(x)$


```python
## Args:
##    x         a value to compute the approximation at it.
##    degree    the term to be stopped at
##    a         the center of approximation, default = 0
## Returns:
##    y       approximated value of the exponential function
def exp_taylor_expansion(x, degree, a=0):
  y = 0
  for n in range(degree+1):
    y = y + (x ** n) / math.factorial(n)
  return y

  ## set x at a certain value near the center
x = 2 #
print(f'The true exp ({x}) is: {np.exp(x)}')
for n in range(0,8):
  y = exp_taylor_expansion(x, n)
  print(f'\n_______________________________\nUsing {n}-term, our result is {y}, \nabs_error = {np.abs(y-np.exp(x))}, relative error = {np.abs(y-np.exp(x))/np.exp(x)}')
```

## compare $x$ values [1, 10, 100, 1000] (**close** to & **far** from **$a$**)

```python
## set x at several points
X = [1, 10, 100, 1000]
for x in X:
  ## use the best approximated value (with lowest error) at n=2
  y = exp_taylor_expansion(x, 7)  # الدرجة 7 هي ادق لانها اكبر عشان كذا استخدمتها
  print(f'The true exp ({x}) is: {np.exp(x)}, using {n}-term is {y}, abs_error is = {np.abs(y-np.exp(x))}')
  # 1000 طلع ناتج كبير فوق الحد فطلع overflow
  # كلما كانت الاكس اصغر كان ادق او اننا بحاجة لزيادة الدرجة وزيادة سلسلة تايلور حتى نقترب اكثر
```

# Approximate polynomial $f(x)$ using `approximate_taylor_polynomial`
# ⚓ second **Function:**: $y = 5x^4 + 3.5x^3 - 5.9x^2 + 4x - 15.2$


**Taylor Series Expansion Around $a = 0$:**

The Taylor series expansion of a polynomial is the polynomial itself. Therefore, the Taylor series expansion of the given function is:

$$y = 5x^4 + 3.5x^3 - 5.9x^2 + 4x - 15.2$$

```python
from scipy.interpolate import approximate_taylor_polynomial
```

```python
## Define your function, where it takes x and return y
def my_func(x):
  y = 5* x**4 +3.5*x**3 -5.9 *x**2 +4*x -15.2
  return y
```

```python
x = np.linspace(-np.pi, np.pi, 200)  ## to get an array with evenly spaced values. (size = 200)
y = np.zeros(len(x))  ## initialize all y for every x above to 0 to prepare the sum (size= same as x)
plt.figure(figsize = (7,6)) ## setting the size of the plot

## link the iterator with its corresponding label (using zip),
## then, loop throguh them (using for)
for n in range(8):
    ## compute the current order approcximation by accumulating the current degree term
    taylor_app = approximate_taylor_polynomial(my_func, 0, n, 1)
    ## plot the current approximation
    plt.plot(x, taylor_app(x), label=f'Degree {n}')

## plot the analytic (actual) function by using function
plt.plot(x, my_func(x), 'k', label = 'Analytic value')

## configure the plot
plt.grid()      # show the grid lines
plt.title('Taylor Series Approximations of Various Orders')  # add plot title
plt.xlabel('x') # add x-axis label
plt.ylabel('y') # add x-axis label
plt.legend()    # show the legend (as we have a label for each line)
plt.show()
```

> We see that once we increase the degree we come closer to the true value::


# Comparing analytical and approximated value of $f(x)$


```python
## Args:
## x         a value to compute the approximation at it.
## degree    the term to be stopped at
## a         the center of approximation, default = 0
## Returns: y  approximated value of the exponential function
def taylor_expansion(x, degree, a=0):
  for n in range(degree+1):
    taylor_app = approximate_taylor_polynomial(my_func, 0, n, 1)
  return taylor_app(x)
  ## set x at a 0.1 value near the center
x = 0.1
func=5* x**4 +3.5*x**3 -5.9 *x**2 +4*x -15.2
print(f'The true func ({x}) is: {func}')

for n in range(0,8):
  y = taylor_expansion(x, n)
  print(f'\n_______________________________\nUsing {n}-term, our result is {y}, \nabs_error = {np.abs(y-func)}, relative error = {np.abs(y-func)/func}')
```

## compare $x$ values [1, 10, 100, 1000] (**close** to & **away** from **$a$**)

```python
## set x at several points
X = [1, 10, 100, 1000]
for x in X:
  ## use the best approximated value (with lowest error) at n=2
  y = taylor_expansion(x, 7)  # الدرجة 7 هي ادق لانها اكبر عشان كذا استخدمتها

  func=5* x**4 +3.5*x**3 -5.9 *x**2 +4*x -15.2
  print(f'The true func ({x}) is: {func}, using {n}-term is {y}, abs_error is = {np.abs(y-func)}')
  #  كلما كانت الاكس اصغر كان ادق , اننا بحاجة لزيادة الدرجة وزيادة سلسلة تايلور حتى نقترب اكثر وتكون ادق
```

### conclusion
As expected,


> increasing the degree of the Taylor polynomial generally improves the accuracy of the approximation


 **in the both two examples** particularly evident when evaluating the function at points closer to the center of the expansion, which in this case is a = 0.
as we move further away from the center point, the accuracy of the approximation diminishes. This is demonstrated by the increasing absolute error for larger values of x. To ***maintain accuracy*** for larger values, it may be necessary to either **increase the degree** of the Taylor polynomial or consider using a **different center point.**
